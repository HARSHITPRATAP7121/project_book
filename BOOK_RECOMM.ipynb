```python
# Lets store users who had at least rated more than 200 books
x = ratings['user_id'].value_counts() > 200
x[x].shape
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
# Now join ratings with books

ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.head()
```

```python
ratings_with_books.shape
```

```python

number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
number_rating.rename(columns={'rating':'num_of_rating'},inplace=True)
```

```python
number_rating.head()
```

```python
final = ratings_with_books.merge(number_rating, on='title')
```

```python
final.head()
```

```python
final.shape
```

```python
final = final[final['num_of_rating'] >= 80]
```

```python
final.head()
```

```python
final.shape
```

```python
final.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final.shape
```

```python
# Lets create a pivot table
book_pivot = final.pivot_table(columns='user_id', index='title', values= 'rating')
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
type(book_sparse)
```

```python
# Now import our clustering algoritm which is Nearest Neighbors this is an unsupervised ml algo
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

```python
model.fit(book_sparse)
```

```python

distance, suggestion = model.kneighbors(book_pivot.iloc[241,:].values.reshape(1,-1), n_neighbors=6 )
```

```python

def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6 )
    
    for i in range(len(suggestion)):
            books = book_pivot.index[suggestion[i]]
            for j in books:
                if j == book_name:
                    print(f"You searched '{book_name}'\n")
                    print("The suggestion books are: \n")
                else:
                    print(j)
```

```python
book_name = "Harry Potter and the Chamber of Secrets (Book 2)"
recommend_book(book_name)
```
